```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df.info()
```

```python
sex=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
df = pd.concat([df, sex], axis=1)
```

```python
df.head()
```

```python
sns.countplot(x= 'Sex', data=df)
plt.show()
```

```python
sns.countplot(x='Pclass', data=df, hue='Sex')
plt.show()
```

```python
 def male_female_child(passenger):
        age, sex = passenger
        
        if age < 16:
            return "child"
        else:
            return sex
```

```python
df['person'] = df[['Age', 'Sex']].apply(male_female_child, axis=1)
```

```python
df[0:10]
```

```python
sns.countplot(x='Pclass', data=df, hue='person')
plt.show()
```

```python
df["Age"].hist(bins=70)
plt.show()
```

```python
fig = sns.FacetGrid(df, hue="Sex", aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)

oldest = df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
plt.show()
```

```python
fig = sns.FacetGrid(df, hue="person", aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)

oldest = df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
plt.show()
```

```python
fig = sns.FacetGrid(df, hue="Pclass", aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)

oldest = df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
plt.show()
```

# Who was alone and who was with family

```python
df.head()
```

```python
df['Alone'] = df.SibSp + df.Parch
```

```python
df["Alone"]
```

```python
df["Alone"].loc[df['Alone'] > 0] = 'With Family'
df['Alone'].loc[df['Alone'] == 0] = "Alone"
```

```python
df.head()
```

```python
sns.countplot(x='Alone', data=df, palette='Blues')
plt.show()
```

```python
df["Survivor"] = df.Survived.map({0:'no',1:'yes'})

sns.countplot(x='Survivor', data=df, palette='Set1')
plt.show()
```

```python
sns.factorplot('Pclass','Survived', data=df, hue='person')
plt.show()
```

```python
sns.lmplot('Age','Survived',data=df)
plt.show()
```

```python
sns.lmplot('Age','Survived',data=df,hue='Pclass')
plt.show()
```

```python
generations =  [10, 20, 40, 60, 80]
sns.lmplot('Age','Survived',data=df,hue='Pclass', palette='winter',x_bins= generations)
plt.show()
```

```python
sns.lmplot('Age', 'Survived', hue='Sex', data=df, palette='winter', x_bins=generations)
plt.show()
```
